# Harmonic Oscillator - Basic

One mass, one spring, single DOF, no damping, no external force
The equation of motion for this sistem is:

$m\ddot x + k x = 0$

With the solution:

$x(t) = a \cdot sin(\omega t + \phi) $

where a and $\phi$ are derived from initial conditions, and $\omega = \sqrt \frac{k}{m} $

This analitical solution can be plotted immidiately:

```python
# Import
import numpy as np # vectors and matrices function
import matplotlib.pyplot as plt # plotting figures
```

```python
a = 1 # initial conditions
phi = 0 # initial conditions
f = 1 #[Hz] the frequency is determined by ration between k and m
omega = 2*np.pi*f

t_an = np.linspace(0,10,1000)
x_an = a * np.sin(omega*t_an + phi)

#plt.figure
fig, ax = plt.subplots()
ax.plot(t_an,x_an)
ax.set(xlabel='time [sec]', ylabel='$sin(\omega t)$',
       title='Basic Harmonic oscillation')

ax.grid()
#hold on
```

## Numerical Integration

There of course no neet to solve this numerically. But it would be useful to start checking numerical simulation on the simple, analitical case. <br>
System parameters: <br>
Normalized mass: m = 1 <br>
In the previous section we chose f=1[Hz] <br>
This resukts in k: $\omega = 2 \pi f = \sqrt \frac{k}{m} \Rightarrow (2 \pi)^2 = k$

```python
m = 1
k = 4*np.pi*np.pi
```

Initial conditions: <br>
$x_0 = 0$ <br>
$\dot x = \omega \cdot sin(\omega t)$ <br>
$f=1[Hz] \Rightarrow \omega = 2\pi \Rightarrow \dot x _0 = 2 \pi $

```python
x0 = 0
xdot0 = 2*np.pi
```

### The simplest and the most basic integration is the simple summation:
$y_{i+1} = y_i + \dot y_i \cdot \Delta t$

```python
fig, ax = plt.subplots()
dt = 0.01 #integration interval [sec]
#initialization
t = 0;
x = x0;
xdot = xdot0;
#iterative solution
for ind in range(1, 1000):
    t = t + dt
    xdotdot = -k/m*x
    xdot = xdot + xdotdot * dt
    x = x + xdot * dt
    ax.plot(t,x,'r.')

ax.plot(t_an, x_an, label = 'Analitical')
ax.set(xlabel='time [sec]', ylabel='$sin(\omega t)$',
       title='Basic Harmonic oscillation')
```
